## Retrieve the data on popularity and genre

```python
def get_token():
    auth_string = client_id5 + ":" + client_secret5
    auth_bytes = auth_string.encode("utf-8")
    auth_base64 = str(base64.b64encode(auth_bytes), "utf-8")
    
    url = "https://accounts.spotify.com/api/token"
    headers = {
        "Authorization": "Basic " + auth_base64,
        "Content-Type": "application/x-www-form-urlencoded"
    }
    data = {"grant_type": "client_credentials"}
    result = post(url, headers=headers, data=data)
    json_result = json.loads(result.content)
    token = json_result["access_token"]
    return token
```

```python
def get_auth_header(token):
    return {"Authorization": "Bearer " + token}
```

```python
def search_for_track_popularity(token, track_name):
    url = "https://api.spotify.com/v1/search"
    headers = get_auth_header(token)
    query = f"?q={track_name}&type=track&limit=1"
    
    query_url = url + query
    result = get(query_url, headers=headers)
    json_result = json.loads(result.content)
    
    try:
        return json_result['tracks']['items'][0]['popularity']
    except:
        return None
```

```python
def search_for_artist_genre(token, artist_name):
    url = "https://api.spotify.com/v1/search"
    headers = get_auth_header(token)
    query = f"?q={artist_name}&type=artist&limit=1"
    
    query_url = url + query
    result = get(query_url, headers=headers)
    json_result = json.loads(result.content)
    
    try:
        # if 'error' in json_result.keys():
        #     return None
        # else:
        return json_result['artists']['items'][0]['genres']
    except:
        return None
```

## Overview of the Spotify data

```python
filelist = ['data', 'data2', 'data3', 'data4', 'data5', 'data6']

def combine_file(filelist):
    dflist = []
    for file in filelist:
        file = pd.read_json(f'{file}.json', lines=True)
        dflist.append(file)
    merged_df = pd.concat(dflist, ignore_index=True)
    return merged_df
    
spotify_data = combine_file(filelist)
spotify_data
```

```python
spotify_data.to_csv('spotify_data.csv')
```
